```python
#library
import pandas as pd
import numpy as np
```

```python
player_df = pd.read_csv("data.csv")
player_df
```

```python
numcols = ['Overall', 'Crossing','Finishing',  'ShortPassing',  'Dribbling','LongPassing', 'BallControl', 'Acceleration','SprintSpeed', 'Agility',  'Stamina','Volleys','FKAccuracy','Reactions','Balance','ShotPower','Strength','LongShots','Aggression','Interceptions']
catcols = ['Preferred Foot','Position','Body Type','Nationality','Weak Foot']
```

```python
len(numcols)
```

```python
print(len(player_df.columns))
player_df.columns
```

```python
player_df = player_df[numcols+catcols]
player_df
```

```python
player_df[catcols]
```

```python
pd.get_dummies(player_df[catcols])
```

```python
traindf = pd.concat([player_df[numcols], pd.get_dummies(player_df[catcols])],axis=1)
features = traindf.columns

traindf = traindf.dropna()
```

```python
traindf = pd.DataFrame(traindf,columns=features)
traindf
```

```python
[print(i) for i in traindf.columns.tolist()]
```

```python
y = traindf['Overall']>=87
X = traindf.copy()
del X['Overall']
```

```python
X.head()
```

```python
len(X.columns)
```

```python
feature_name = list(X.columns)
print(len(feature_name))
print(feature_name)
# no of maximum features we need to select
num_feats=30
```

```python
def cor_selector(X, y,num_feats):
    cor_list = []
    feature_name = X.columns.tolist()
    # calculate the correlation with y for each feature
    for i in X.columns.tolist():
        cor = np.corrcoef(X[i], y)[0, 1]
        print(cor)
        cor_list.append(cor)
    # replace NaN with 0
    cor_list = [0 if np.isnan(i) else i for i in cor_list]
    print('mean cor_list :', np.mean(cor_list))
    
#     print('corlist:', cor_list)
    print("len_corelist: ",len(cor_list))
    # feature name
    print("(np.abs(cor_list)): ", (np.abs(cor_list))[-num_feats:])
    print("(np.abs(cor_list)) sort top 30: ", sorted(np.abs(cor_list))[-num_feats:])
    print('np.argsort(np.abs(cor_list))[-num_feats:]]: ', np.argsort(np.abs(cor_list))[-num_feats:])
#     print('check: ',np.argsort(np.abs(cor_list))[-num_feats:])
    
    cor_feature = X.iloc[:,np.argsort(np.abs(cor_list))[-num_feats:]].columns.tolist()
    # feature selection? 0 for not select, 1 for select
    cor_support = [True if i in cor_feature else False for i in feature_name]
    return cor_support, cor_feature
cor_support, cor_feature = cor_selector(X, y,num_feats)
print(str(len(cor_feature)), 'selected features')
```
